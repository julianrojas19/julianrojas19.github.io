```python
from scipy.spatial.distance import mahalanobis
from scipy.stats import chi2
import scipy.stats as stats
```

```python
airbnb_mahalanobis=airbnb[['price','number_of_reviews','calculated_host_listings_count','availability_365']]
```

```python
mahal_distances = []

for row in airbnb_mahalanobis.to_numpy():
  mahal_distances.append(mahalanobis(row, airbnb_mahalanobis.mean(), np.linalg.inv(np.cov(airbnb_mahalanobis.values.T))))
```

```python
airbnb['mahal_distances'] = mahal_distances
airbnb['p_value'] = 1 -stats.chi2.cdf(airbnb['mahal_distances'], 13-1)
airbnb.head()
```
